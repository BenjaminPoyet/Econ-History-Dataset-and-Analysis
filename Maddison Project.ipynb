```python
GDP_Levels_melt = pd.melt(GDP_Levels, id_vars=["Country"], var_name="year")
GDP_Levels_melt = GDP_Levels_melt[GDP_Levels_melt["Country"].str.contains('World Total')]
GDP_Levels_melt = GDP_Levels_melt.pivot(index='year', columns='Country', values='value')
GDP_Levels_melt.plot()
```

```python
GDP_Levels_melt=GDP_Levels_melt.reset_index()
GDP_Levels_melt.head()
```

```python
GDP_Levels_melt=GDP_Levels_melt.dropna()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
def sigmoid(x, Beta_1, Beta_2):
     y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
     return y
plt.figure(figsize=(8,5))
x_data, y_data = (GDP_Levels_melt["year"].values, GDP_Levels_melt["World Total"].values)
plt.plot(x_data, y_data, 'ro')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

```python
beta_1 = 0.10
beta_2 = 1990.0

#logistic function
Y_pred = sigmoid(x_data, beta_1 , beta_2)

#plot initial prediction against datapoints
plt.plot(x_data, Y_pred*20000000.)
plt.plot(x_data, y_data, 'ro')
```

```python
xdata =x_data/max(x_data)
ydata =y_data/max(y_data)
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, xdata, ydata)
#print the final parameters
print(" beta_1 = %f, beta_2 = %f" % (popt[0], popt[1]))
```

```python
x = np.linspace(1, 2008, 700)
x = x/max(x)
plt.figure(figsize=(8,5))
y = sigmoid(x, *popt)
plt.plot(xdata, ydata, 'ro', label='data')
plt.plot(x,y, linewidth=3.0, label='fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

```python
#Evaluation of the model
# split data into train/test
msk = np.random.rand(len(GDP_Levels_melt)) < 0.8
train_x = xdata[msk]
test_x = xdata[~msk]
train_y = ydata[msk]
test_y = ydata[~msk]

# build the model using train set
popt, pcov = curve_fit(sigmoid, train_x, train_y)

# predict using test set
y_hat = sigmoid(test_x, *popt)

# evaluation
print("Mean absolute error: %.2f" % np.mean(np.absolute(y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_hat - test_y) ** 2))
from sklearn.metrics import r2_score
print("R2-score: %.2f" % r2_score(y_hat , test_y) )
```
